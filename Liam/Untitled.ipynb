## Upload data to database

```python
for row in ballarat.iterrows():
    dicko = row[1].to_dict()
    dicko['location'] = {'type':'Point', 'coordinates':[dicko['LONG'], dicko['LAT']]}
    dicko.__delitem__('LAT')
    dicko.__delitem__('LONG')
    db.sportsgrounds.insert_one(dicko)
```
